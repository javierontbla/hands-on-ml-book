A Support Vector Machine (SVM) model is capable of performing **linear and nonlinear classification, regression and even outlier detection**. SVMs are well suited for small and medium sized datasets. SVMs are sensitive to feature scaling.

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.svm import LinearSVC, SVC, LinearSVR
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.datasets import make_moons
```

```python
# we are going to load the iris dataset, scale feature (for the SVM model) and train it.
iris = datasets.load_iris()
X = iris["data"][:, (2, 3)] # petal length and width
y = (iris["target"] == 2).astype(np.float64) # get only Iris Virginica and convert boolean to 0s and 1s

# pipeline to prepare data
svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC(C=1, loss="hinge"))
])

# train model
svm_clf.fit(X, y)
```

```python
# plot support vectors with the help of DecisionBoundaryDisplay from sklearn
# and we can see that the data is linearly separable
plt.scatter(X[:, 0], X[:, 1], c=y, marker='x')
ax = plt.gca() # get current axes
DecisionBoundaryDisplay.from_estimator(
    svm_clf,
    X,
    ax=ax,
    grid_resolution=50,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
)
    
plt.show()
```

This example was with linearly separable data, however this is not always the case. So, to perform this on nonlinear data, we can add more features, such as polynomial features to create a linearly separable data. 

```python
X_moon, y_moon = make_moons(n_samples=100, noise=0.15) # nonlinear data
poly_svm_clf = Pipeline([
    ("poly_features", PolynomialFeatures(degree=3)),
    ("scaler", StandardScaler()),
    ("svm_clf", LinearSVC(C=10, loss="hinge"))
])
poly_svm_clf.fit(X_moon, y_moon) # train it

# plot data and boundaries
ax = plt.gca() # get current axes
DecisionBoundaryDisplay.from_estimator(
    poly_svm_clf,
    X_moon,
    ax=ax,
    grid_resolution=50,
    plot_method="contour",
    colors="g",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
)
plt.scatter(X_moon[:, 0], X_moon[:, 1], c="k", marker='o')
```

```python
# to improve performance, we can use the "kernel trick" to get the same result 
# as if we had added polynomial features, without actually doing it
poly_kernel_svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
])
poly_kernel_svm_clf.fit(X_moon, y_moon)

# plot data and boundaries
ax = plt.gca() # get current axes
DecisionBoundaryDisplay.from_estimator(
    poly_kernel_svm_clf,
    X_moon,
    ax=ax,
    grid_resolution=50,
    plot_method="contour",
    colors="g",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
)
plt.scatter(X_moon[:, 0], X_moon[:, 1], c="k", marker='o')
```

```python
# we can also use SVM for regression tasks
m = 100
eps = 1.5

X_rand = 2 * np.random.rand(m, 1)
y_rand = 4 + 3 * X_rand + np.random.randn(m, 1)

svm_reg = LinearSVR(epsilon=eps) # with epsilon we define the width of the "street"
svm_reg.fit(X_rand, y_rand)

y_line = svm_reg.intercept_[0] + svm_reg.coef_[0] * X_rand
y_tol_01 = (svm_reg.intercept_[0] - eps) + svm_reg.coef_[0] * X_rand
y_tol_02 = (svm_reg.intercept_[0] + eps) + svm_reg.coef_[0] * X_rand

plt.scatter(X_rand, y_rand, c="g")
plt.plot(X_rand, y_line, "r-")
plt.plot(X_rand, y_tol_01, linestyle='--', color='black')
plt.plot(X_rand, y_tol_02, linestyle='--', color='black')
plt.show()
```
