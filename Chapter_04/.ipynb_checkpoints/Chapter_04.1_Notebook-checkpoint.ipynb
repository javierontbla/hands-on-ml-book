```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np

# set data points from example
X = [2, 5, 3]
y = [2, 4, 5]

# plot data points and x and y limits
plt.scatter(X, y, c='#4ad66d')
plt.xlim((0, 6))
plt.ylim((0, 6))

# set initial values of bias, eta and iterations
bias = 0
eta = 0.1
n_iterations = 1000

# compute gradient descent
for iteration in range(n_iterations):
    # calculate gradients
    gradients = (1/3) * ((-2 * (2 - (bias + (0.4 * 2)))) + (-2 * (5 - (bias + (0.4 * 3)))) + (-2 * (4 - (bias + (0.4 * 5)))))
    bias = bias - (eta * gradients)

    # plot bias change
    y_predict = []
    for x in X:
        y_predict.append(bias + (0.4 * x))
    plt.plot(X, y_predict, color='#00a8e8')

    # calculate MSE on every iteration
    errors = []
    for i in range(len(X)):
        errors.append((y[i] - (bias + (0.4 * X[i]))) ** 2)
    mse = (1/3) * np.sum(errors)
    print(f"MSE: {mse} and Bias: {bias}")

plt.show()
```
