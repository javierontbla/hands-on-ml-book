```python
# linear reg - simple
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# change plot's background color
plt.figure(facecolor='#1D1E22')
ax = plt.axes()
ax.set_facecolor('#1D1E22')

figure = plt.gcf()  # get current figure
figure.set_size_inches(8, 6) # set figure's size manually to your full screen (32x18)

# change axis colors
ax.spines[['right', 'top']].set_visible(False)
ax.spines['bottom'].set_color('#f7f7f7')
ax.spines['left'].set_color('#f7f7f7')
ax.xaxis.label.set_color('#f7f7f7')
ax.yaxis.label.set_color('#f7f7f7')
ax.tick_params(axis='x', colors='#f7f7f7')
ax.tick_params(axis='y', colors='#f7f7f7')
ax.set_yticks([])
ax.set_xticks([])

m = 100 # no. de instancias
X = np.random.rand(m, 1) # atributos
y = 2 + 3 * X + np.random.rand(m, 1) # etiquetas (ecuación original: 2 + 3 * x1)

# 4. entrenar modelo
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# 5. conjunto para graficar línea de regresión
X_plot = np.linspace(0, 1, m).reshape(m, 1)
y_plot = lin_reg.predict(X_plot)

plt.scatter(X, y, s=40, c='#18EF62', alpha=1.0)
plt.plot(X_plot, y_plot, color='#3AAAFF', linewidth=4)
#plt.savefig('lin-reg-2d.png', bbox_inches='tight', transparent=True)
plt.show()
```

```python
# linear reg - multiple
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
import sklearn.linear_model
from mpl_toolkits.mplot3d import Axes3D

X_train = np.random.rand(m * 2).reshape(100,2) * 10
y_train = (X_train[:, 0]**2)+(X_train[:, 1]**2)

fig = plt.figure()
fig.set_size_inches(18, 12)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train[:,0], X_train[:,1], y_train, s=180, marker='.', color='#18EF62')

# change plot's background color
ax.set_facecolor('#1D1E22')
#ax.patch.set_facecolor('None')

model = sklearn.linear_model.LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_train)

coefs = model.coef_
intercept = model.intercept_
xs = np.tile(np.arange(10), (10,1))
ys = np.tile(np.arange(10), (10,1)).T
zs = xs*coefs[0]+ys*coefs[1]+intercept

for axis in [ax.xaxis, ax.yaxis, ax.zaxis]:
    axis.set_ticklabels([])
    axis._axinfo['axisline']['linewidth'] = 1
    axis._axinfo['axisline']['color'] = "#f7f7f7"
    axis._axinfo['grid']['linewidth'] = 0.5
    axis._axinfo['grid']['linestyle'] = "--"
    axis._axinfo['grid']['color'] = "#f7f7f7"
    axis._axinfo['tick']['inward_factor'] = 0.0
    axis._axinfo['tick']['outward_factor'] = 0.0
    axis.set_pane_color('None')

rcParams['xtick.color'] = 'None'
rcParams['ytick.color'] = 'None'
rcParams['axes.labelcolor'] = 'None'
rcParams['axes.edgecolor'] = 'None'

ax.plot_surface(xs,ys,zs, alpha=0.4, color='#3AAAFF')
#plt.savefig('lin-reg-3d.png', bbox_inches='tight', transparent=True)
plt.show()
```
