```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, train_test_split
import matplotlib.pyplot as plt
import numpy as np

# generar conjunto de datos
m = 100 
X = 6 * np.random.rand(m, 1)
y = 2 + X + np.random.rand(m, 1) * 2

# dividir en conjunto de datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape, X_test.shape)

# entrenar modelo
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# conjunto para graficar línea de regresión
X_plot = np.linspace(0, 6, m).reshape(m, 1)
y_plot = lin_reg.predict(X_plot)

# crear gráficos
plt.scatter(X, y, s=20, c='#18EF62', alpha=0.9)
plt.plot(X_plot, y_plot, color='#3AAAFF', linewidth=4)

# evaluar modelo (ECM) con X_train
y_train_pred = lin_reg.predict(X_train)
train_mse = mean_squared_error(y_train, y_train_pred)
print(f"ECM Entrenamiento: {train_mse}")

# evaluar modelo (X_train) con Cross-Validation
cross_scores_train = cross_val_score(lin_reg, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
print(-cross_scores_train)

# evaluar modelo (ECM), pero ahora con X_train
y_test_pred = lin_reg.predict(X_test)
test_mse = mean_squared_error(y_test, y_test_pred)
print(f"ECM Prueba: {test_mse}")

# evaluamos modelo con cross-validation, pero ahora con X_test
cross_scores_test = cross_val_score(lin_reg, X_test, y_test, scoring='neg_mean_squared_error', cv=10)
print(-cross_scores_test)
```
