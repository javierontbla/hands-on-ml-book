```python
import numpy as np
import matplotlib.pyplot as plt

def gradiente_descendente_batch():
    # 1. Crear conjunto de datos (y = ϴ0 + ϴ1 * x1)
    m = 100
    X = np.random.rand(m, 1)
    y = 2 + 3 * X + np.random.rand(m, 1)
    
    # 2. Graficar conjunto de datos
    plt.scatter(X, y)
    plt.show()
    
    # 3. Establecer tasa de aprendizaje, no. de iteraciones y parámetros aleatorios
    eta = 0.1
    no_iterations = 1000
    theta = np.random.rand(2, 1)

    # 4. Agregar x0 a X
    X = np.c_[np.ones((100, 1)), X]
    
    # 5. Bucle "for" para implementar el algoritmo
    for iteration in range(0, no_iterations):
        gradient = 2 / m * X.T.dot(X.dot(theta) - y)
        theta = theta - (eta * gradient)
    
    return theta

print(gradiente_descendente_batch())
```
