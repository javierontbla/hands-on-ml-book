```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# generar conjunto de datos a partir de función cuadrática
m = 100 # no. de instancias
X = 6 * np.random.rand(m, 1) - 3 # X = 6 * ruido - 3
y = 2 + X + 0.5 * X**2 + np.random.rand(m, 1) # y = 2 + 1x + 0.5x^2 + ruido

# generar atributos polinomiales con PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# imprimir una instancia aleatoria para ver la diferencia
print(f"Original: {X[0]} || Polinomial: {X_poly[0]}")

# entrenamos un modelo de regresión lineal con X_poly
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

# imprimir término bias y coeficientes
print(f"y = {lin_reg.intercept_[0]} + {lin_reg.coef_[0][0]}x +  {lin_reg.coef_[0][1]}x^2")

# valores de X y y para graficar la curva
X_plot = np.linspace(-3, 3, m).reshape(m, 1)
X_plot_poly = poly_features.transform(X_plot)
y_plot = lin_reg.predict(X_plot_poly) # predecir con atributos polinomiales

# graficar conjunto de datos
plt.scatter(X, y, s=10, c='#000000')
plt.plot(X_plot, y_plot, color='r') # graficar con X_plot original
plt.show()
```
