# Chapter 03 - Classification

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, precision_recall_curve, roc_curve, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import sys
sys.path.append('/usr/local/lib/python2.7/site-packages')
```

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
```

```python
X, y = mnist["data"], mnist["target"]
rand_num = X.iloc[0]
rand_num_img = rand_num.values.reshape(28, 28)

plt.imshow(rand_num_img, cmap="bone")
plt.axis("off")
plt.show()
```

```python
# because all "y" elements are strings, we need to convert them to int
y = y.astype(np.uint8)

# now we split dataset into train and test set
# the MNIST dataset is already shuffled
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]

# now we train a Binary Classifier that detects 5s
y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5)

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5)
```

```python
# our classifier now classifies 5s correctly
sgd_clf.predict([rand_num])
```

### Measure Classifier's Performance

```python
# using cross-validation to measure accuracy
# (accuracy is not the best performance measure for classifiers)
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

```python
# now we use a confusion matrix
# [ T.N.   F.P ] True Negatives & False Positives
# [ F.N.   T.P ] False Negatives & True Positives
# a perfect classifier would have only true positives and true negatives

# this method returns the predictions made, instead of scores
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3)
confusion_matrix(y_train_5, y_train_pred)
```

```python
# we can also measure precision and recall
precision = precision_score(y_train_5, y_train_pred)
recall = recall_score(y_train_5, y_train_pred)

print(f"Precision: {precision} and Recall: {recall}")
```

```python
# and something called the F1 Score combines precision and recall
f1_score_var = f1_score(y_train_5, y_train_pred)
f1_score_var
```

```python
# the classifier works with a threshold
# for example, if the decision function score is greater than 2500
# then the number is a 5, otherwise is not a 5
y_score = sgd_clf.decision_function([rand_num])

# we now change the threshold (default is 0)
threshold = 8000
y_some_rand_num = (y_score > threshold)

print(f"Score is: {int(y_score[0])} but threshold is {threshold}")
```

```python
# but how do we decide which threshold to use?
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3, method="decision_function")

# now that we have the decision functions scores of all instances
# we can plot precision/recall trade-off 
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle("Precision vs Recall")

ax1.plot(thresholds, precisions[:-1], "b--", label="Precision")
ax1.plot(thresholds, recalls[:-1], "g-", label="Recall")
ax1.set_xlim([-50000, 50000])
ax1.legend()
ax1.grid()

ax2.plot(recalls, precisions, "k--")
ax2.grid()
```

```python
# to get the threshold where precision is >= 90%
# we get the index where precisions is >= 90% in the thresholds array
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]

# now, instead of using predict, we use the prediction scores 
# and set to True the ones that are greater then the current threshold
y_train_pred_90 = (y_scores >= threshold_90_precision) # the ones ABOVE the thresghold become True

new_precision = precision_score(y_train_5, y_train_pred_90)
new_recall = recall_score(y_train_5, y_train_pred_90)

print(f"The new precision score is: {new_precision} while the recall score is: {new_recall}")
```

```python
# now let's measure the performance with the ROC curve
# a good classifier stays as far away from the dotted black line as possible
fpr, tpr, thresholds_roc =roc_curve(y_train_5, y_scores)
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0, 1], [0, 1], "k--")
plt.grid()
plt.show()
```

```python
# we can also compute the Area Under the Curve (AUC) score
roc_auc_score(y_train_5, y_scores)
```

```python
forest_clf = RandomForestClassifier(random_state=42)
forest_clf.fit(X_train, y_train_5)
```

```python
forest_clf.predict_proba([rand_num]) # 2% it is NOT a 5, 98% it is a 5
```

```python
# plot the ROC curve of both classifiers to see the difference
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train_5, cv=3, method="predict_proba")
y_scores_forest = y_probas_forest[:, 1]

fpr_forest, tpr_forest, thresholds_forest = roc_curve(y_train_5, y_scores_forest)
plt.plot(fpr, tpr, "b:", label="SGD")
plt.plot(fpr_forest, tpr_forest, "k--", label="Random Forest")
plt.legend(loc="lower right")
plt.show()
```

```python
roc_auc_score(y_train_5, y_scores_forest)
```

```python
# let's measure precision and recall from this new Forest classifier
# (this classifier is way better than the SGDClassifier)
precisions_forest, recalls_forest, thresholds_forest = precision_recall_curve(y_train_5, y_scores_forest)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle("Precision vs Recall (Random Forest)")

ax1.plot(thresholds_forest, precisions_forest[:-1], "b--", label="Precision")
ax1.plot(thresholds_forest, recalls_forest[:-1], "g-", label="Recall")
ax1.legend()
ax1.grid()

ax2.plot(recalls_forest, precisions_forest, "k--")
ax2.grid()
```

```python
y_train_pred_forest = (y_scores_forest >= 0.6)
precision_forest = precision_score(y_train_5, y_train_pred_forest)
recall_forest = recall_score(y_train_5, y_train_pred_forest)

print(f"Precision from Random Forest is: {precision_forest} and Recall is: {recall_forest}")
```

### Multiclass Classification

```python
# we can also train multiclass classifiers
# using OvR = One vs the Rest
# or OvO = One vs One approaches
svm_clf = SVC()
svm_clf.fit(X_train, y_train)
svm_clf.predict([rand_num])
```

```python
# we can get the decision scores (the highest is the winner)
rand_num_scores = svm_clf.decision_function([rand_num])
rand_num_scores
```

```python
# and we can get the classes 
svm_clf.classes_
```

```python
# we can also measure this classifier
# first with the confusion matrix
y_train_pred = cross_val_predict(svm_clf, X_train, y_train, cv=3)
conf_matrix = confusion_matrix(y_train, y_train_pred)

plt.matshow(conf_matrix, cmap=plt.cm.gray)
plt.show()
```

```python
# x-axis = predicted classes
# y-axis = actual classes
class_sums = conf_matrix.sum(axis=1, keepdims=True)
norm_conf_matrix = conf_matrix / class_sums
np.fill_diagonal(norm_conf_matrix, 0)
plt.matshow(norm_conf_matrix, cmap=plt.cm.gray)
plt.show()
```

```python
# we can also train a multilabel classifier
# in this case to detect two characteristics:
# big and odd numbers
y_train_large = (y_train >= 7) # large, means greater or equal to 7 (7, 8, 9)
y_train_odd = (y_train % 2 == 1) # get odd numbers
y_multilabel = np.c_[y_train_large, y_train_odd]

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_multilabel)
```

```python
knn_clf.predict([rand_num])
```

```python
# evaluate the a multilabel classifier with the F1 score
y_train_knn_pred = cross_val_predict(knn_clf, X_train, y_multilabel, cv=3)
# macro means that all labels are equally important, all weights are the same
f1_score(y_multilabel, y_train_knn_pred, average="macro") 
```
